```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

from src.experiment import Experiment
```

```python
df_all = Experiment.df_experiments(Experiment.read_experiments('1/109bonds'), device='AerSimulator', local_search_doe='fast')
```

```python
df = Experiment.filter_experiments(df_all, ansatz_params_reps=2)

_, axes = plt.subplots(4,5, figsize=(14,8))
axes=axes.T.flatten()
for x in range(8):
    ds = df.reset_index().loc[x]
    # sns.lineplot(ds['step3_monitor_iter_gtheta'], label='step3 cvar (sim)', ax=axes[2*x])
    # sns.lineplot(ds['step3_monitor_iter_best_fx'], label='step3 (sim) best')
    sns.lineplot(ds['step3_iter_best_fx'], label='step3 best (sim)', ax=axes[2*x])
    sns.lineplot(x=[0,len(ds['step3_iter_best_fx'])], y=[ds['refvalue'], ds['refvalue']], label='refval (cplex)', c='r', ax=axes[2*x])
    sns.lineplot(ds['step4_iter_best_fx'], label='step4 best (postprocessing fast) 3x', ax=axes[2*x])
    axes[2*x].set_ylim([94,120])
    axes[2*x].set_xlabel('iter')
    axes[2*x].set_ylabel('objective')
    axes[2*x].set_xticks(np.arange(0, len(ds['step3_iter_best_fx'])+1, ds['step3_num_thetas']))
    if x != 2:
        axes[2*x].get_legend().remove()
    
    axes[2*x+1].plot(np.array(ds.step4_fx_evals))
    axes[2*x+1].set_xlabel('iter')
    axes[2*x+1].set_ylabel('fx_evals_per_shot')
    axes[2*x+1].set_xticks(np.arange(0, len(ds['step3_iter_best_fx'])+1, ds['step3_num_thetas']))

plt.suptitle('TwoLocal 2 reps')
```
